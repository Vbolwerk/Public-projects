### Intro

This project focuses on airport delays. Specifically, the aim of the project is to identify the characteristics of an airport which lead to delays. The project is thus looking for operational factors that are directly correlated to airport delays. Additionally, this project aims to find clusters of airports, that are similar with regards to characteristics such as region or size, to see which kind of airports struggle most with delays.

```python
import sqlite3
sqlite_db = 'airport_db.sqlite'
conn = sqlite3.connect(sqlite_db)
```

```python
df = pd.read_sql("""SELECT * FROM airports 
                INNER JOIN airport_cancellations 
                ON airports.LocID=airport_cancellations.airport 
                INNER JOIN airport_operations 
                ON airport_cancellations.airport=airport_operations.airport 
                AND airport_cancellations.year=airport_operations.year""", 
                con=conn)
```

```python
df.info()
```

There is no missing data in the SQL database, which is reassuring. The data types seem to make sense (values that are supposed to be numerical are numerical, others are objects).

We are looking at data from the United States only, and we are looking at airports of very different sizes. This carries risk with it of small sample sizes for airports that do not have a lot of flights. Additionally, this means that it is important to not pay too much attention to the absolute values that are in the dataframe (such as the absolute value of departure cancellations).

### Exploratory Data Analysis

The first 14 columns (until the column airport) contain characteristics about each airport. Among these characteristics is airport name, its facility type, and the region and location of the airport. After that, the dataframe contains information about the flights to and from these airports. All of this data is numerical, and most of them are not absolute values (so they are either averages or percentages). However, there are also some absolute values (such as departure cancellations), which are absolute values and need to be scaled. 

```python
pd.set_option('display.max_columns', 500)
df.head()
```

```python
df['avg departure cancellations']= df['Departure Cancellations']/ df['departures for metric computation']
df['avg departure diversions']= df['Departure Diversions']/ df['departures for metric computation']
df['avg arrival cancellations']= df['Arrival Cancellations']/ df['arrivals for metric computation']
df['avg arrival diversions']= df['Arrival Diversions']/ df['arrivals for metric computation']
```

#### Are there any duplicate values? 

```python
df = df.drop_duplicates()
```

```python
len(df)
```

### Data Mining

#### Creating Dummy Variables

When getting dummies, dummies for every city, county or state would give us too many features. That is why they are excluded from the dummy variable selection. Facility type are all airports and all airports have boundary data available, so that is also excluded from the dummy selection.

```python
dummie_df = df.loc[:,['FAA REGION', 'AP Type']]
```

```python
dummies = pd.get_dummies(dummie_df)
```

#### Creating a dataframe for the features

```python
num = df.loc[:, 'percent on-time gate departures':]
```

```python
features = num.join(dummies)
```

#### Finding correlations in the data

```python
corr = features.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
plt.show()
```

It is important to find out what the correlations are before PCA, so that it becomes apparent what the main features in the PCA will be. Some of the information about the features will get lost when doing PCA. 

The heatmap above shows different correlations from our dataset, where dark red displays a strong positive correlation, and dark blue displays a strong negative correlation. The relationships seem to make sense. There is for example a strong correlation between the average departures that are cancelled, and the average arrivals that are cancelled. It is also interesting to see that the region AEA (East coast of the US) has positive correlations with average departure delays. This seems to indicate that this region is underperforming in terms of flying on time.

```python
from sklearn.feature_selection import f_regression
```

```python
print 'p-value:'+ str(f_regression(pd.DataFrame(features['average airport departure delay']), pd.DataFrame(features['FAA REGION_AEA']))[1][0])
```

The very small p-value indicates that there is a significant effect of FAA Region AEA on average airport departure delay. This means that airports on the East Coast significantly underperform their peers in terms of departure delays.

## Principal component analysis

```python
# Create a clean data frame 
ap1 = df[['airport','year','Departure Cancellations','Arrival Cancellations']]
print ap1.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
X = pd.DataFrame(StandardScaler().fit_transform(features), columns =features.columns)
```

```python
pca = PCA(n_components=10).fit(X)
```

```python
exp_vars = pca.explained_variance_ratio_
```

```python
plt.plot(exp_vars)
plt.plot(np.cumsum(exp_vars))
plt.title('Explained variance ratio')
plt.show()
```

The most significant elbow in the pca is actually between 1 and 2. Having one component already gives us approximately 0.3 of our explained variance, while the next component gives us 0.1 additional explained variance. We decide to go with four components, because the next elbow is after 4 components.

```python
pca = PCA(n_components=4).fit(X)
```

```python
X_pca = pca.fit_transform(X)
```

```python
i = np.identity(features.shape[1])
```

```python
coef = pca.transform(i)
```

```python
pd.DataFrame(coef, columns=['PC-1', 'PC-2', 'PC-3', 'PC-4'], index=features.columns)
```

```python
plt.scatter(X_pca[:,0], ap1['Departure Cancellations'])
plt.show()
```

```python
plt.scatter(X_pca[:,1], ap1['Departure Cancellations'])
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.gca(alpha = 0.1, projection='3d')
X_plot = X_pca[:,0]
Y_plot = X_pca[:,1]
Z_plot = X_pca[:,2]
surf = ax.scatter(X_plot, Y_plot, Z_plot, alpha=0.2)


plt.show()
```

We can see from the PCA analysis that our principal component does a fairly good job at separating different airports, and explaining the variation between them. When looking at a simple two-dimensional plot with Departure Cancellations, the explanatory power of the principal component seems reasonably good. What is most remarkable in terms of the characteristics of the first principal component is the large part that region can play on the variation of airports. The East coust region (FAA region AEA) is an important predictor of delays, which means that AEA is underperforming when comparing it to its peers. AWP (the western pacific region) on the other hand tends to do quite well when compared to its peers.

When looking at the second principal component, another characteristic that is apparently important in explaining the variance between airports is whether an airport is commercial or private. Public airports in this case tend to underperform its private peers.
