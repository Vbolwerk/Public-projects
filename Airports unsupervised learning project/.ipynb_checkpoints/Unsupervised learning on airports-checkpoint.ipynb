### Intro

This project focuses on airport delays. Specifically, the aim of the project is to identify the characteristics of an airport which lead to delays. The project is thus looking for operational factors that are directly correlated to airport delays. Additionally, this project aims to find clusters of airports, that are similar with regards to characteristics such as region or size, to see which kind of airports struggle most with delays.

## Principal component analysis

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
X = pd.DataFrame(StandardScaler().fit_transform(features), columns =features.columns)
```

```python
pca = PCA(n_components=10).fit(X)
```

```python
exp_vars = pca.explained_variance_ratio_
```

```python
print exp_vars
feature_count = range(1,11)
print feature_count
```

```python
plt.plot(feature_count, exp_vars)
plt.plot(feature_count, np.cumsum(exp_vars))
plt.title('Explained variance ratio')
plt.xlabel('Number of components')
plt.ylabel('Explained \nvariance', rotation = 0)
plt.show()
```

The most significant elbow in the pca is actually between 1 and 2. Having one component already gives us approximately 0.6 of our explained variance, while the next component gives us 0.2 additional explained variance. We decide to go with two components, since the third principal component does not add a lot of explained variance.

```python
pca = PCA(n_components=2).fit(X)
```

```python
X_pca = pd.DataFrame(pca.fit_transform(X), columns=['PC-1', 'PC-2'])
```

```python
i = np.identity(features.shape[1])
```

```python
coef = pca.transform(i)
```

```python
pd.DataFrame(coef, columns=['PC-1', 'PC-2'], index=features.columns)
```

Principal component 1 seems to be positively related to negative attributes (such average airborne delay, average taxi delay and average gate arrival delay) and negatively related to all positive attributes (percent on-time gate departures, for example). This means that an airport with a higher value on principal component 1 tends to perform poorly in terms of delays. 

Principal component 2 is positively related with some positive attributes (such as percent on-time gate departures), but is also positively related to some negative attributes, that mostly have to do with taxi time and taxi delays.
This means that airports with a higher value on principal component 2 tend to have delays mostly because of taxiing times.

```python
plt.scatter(X_pca['PC-1'], X_pca['PC-2'], alpha=0.3)
plt.title('Plot on principal components')
plt.xlabel('Principal component 1')
plt.ylabel('Principal \ncomponent\n    2       ', rotation=0, horizontalalignment='right')
plt.show()
```

We can categorize our airports to certain clusters, so that each airport can get its own strategy of trying to improve delay times, and so that we can find similar airports in our data.

```python
from scipy.cluster.hierarchy import dendrogram, linkage, cophenet, fcluster
from scipy.spatial.distance import pdist
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering

X_pca2=X_pca.copy()
X_pca2["airport"]=df["airport"]
X_pca2["FAA region"]=df["FAA REGION"]
X_pca2["year"]=df["year"]



Z = linkage(X_pca, 'ward')

print "Hierarchical clustering result:"
clusters = fcluster(Z, 30, criterion='distance')
print "Number of cluster:",len(np.unique(clusters))
print "Clusters are defined as:",np.unique(clusters)
X_pca2_h=X_pca2.copy()
X_pca2_h["label"]=clusters

year=df.year.unique()

print "All years"
plt.scatter(X_pca2_h[["PC-1"]],X_pca2_h[["PC-2"]],c= X_pca2_h[["label"]],cmap='prism')
plt.xlabel("Principal component 1")
plt.ylabel("Principal component 2")
plt.show()

print "Over the years"
for y in year:
    print "Year:",y
    xyear=X_pca2_h.loc[X_pca2_h["year"]==y,:]
    xyear.reset_index(inplace=True,drop=True)
    plt.scatter(xyear[["PC-1"]],xyear[["PC-2"]],c=xyear[["label"]],cmap='prism')
    plt.xlabel("Principal component 1")
    plt.ylabel("Principal component 2")
    for i, txt in enumerate(xyear["airport"]):
        (xi,yi)=(xyear.loc[i,"PC-1"],xyear.loc[i,"PC-2"])
        plt.annotate(txt, (xi,yi),fontsize=7)
    plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.gca(alpha = 0.1, projection='3d')
X_plot = X_pca['PC-1']
Y_plot = X_pca['PC-2']
Z_plot = df['departures for metric computation']
surf = ax.scatter(X_plot, Y_plot, Z_plot, alpha=0.2)
plt.xlabel('Principal component 1')
plt.ylabel('Principal component 2')

ax.view_init(30, 270)
plt.show()

fig = plt.figure()
ax = fig.gca(alpha = 0.1, projection='3d')
X_plot = X_pca['PC-1']
Y_plot = X_pca['PC-2']
Z_plot = df['departures for metric computation']
surf = ax.scatter(X_plot, Y_plot, Z_plot, alpha=0.2)
plt.xlabel('Principal component 1')
plt.ylabel('Principal component 2')

ax.view_init(30, 315)
plt.show()

fig = plt.figure()
ax = fig.gca(alpha = 0.1, projection='3d')
X_plot = X_pca['PC-1']
Y_plot = X_pca['PC-2']
Z_plot = df['departures for metric computation']
surf = ax.scatter(X_plot, Y_plot, Z_plot, alpha=0.2)
plt.xlabel('Principal component 1')
plt.ylabel('Principal component 2')

ax.view_init(30, 0)
plt.show()

```

From the 3d-plot, we can tell that the airport size (as measured by number of departures per year) is positively correlated to both principal component 1 and principal component 2. Hence, from this analysis we are able to say that larger on airports tend to have more delays. These delays may be associated with longer taxiing time, as is indicated by the positive correlation with principal component 2.

We can see from the PCA analysis that our principal component does a fairly good job at separating different airports, and explaining the variation between them.
